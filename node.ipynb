```python
import torch

import neural_ode.core as core
import neural_ode.experiments as experiments
import neural_ode.plots as plots

NeuralODEClassifier = core.NeuralODEClassifier
set_seed = core.set_seed
run_experiment = experiments.run_experiment
plot_experiment_report = plots.plot_experiment_report

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", device)
print("Training uses torchdiffeq.odeint_adjoint.")
```

```python
set_seed(42)

model_200 = NeuralODEClassifier(
    latent_dim=2,
    hidden_dim=96,
    encoder_type="parabola",
    integration_end=2.0,
    decoder_mlp=True,
)

result_200 = run_experiment(
    model=model_200,
    train_size=200,
    val_size=4000,
    test_size=10000,
    model_name="200 points",
    epochs=120,
    batch_size=1024,
    lr=3e-3,
    weight_decay=1e-5,
    grad_clip_norm=1.0,
    use_cosine_scheduler=True,
    device=device,
)

plot_experiment_report(result_200, device=device, latent_trajectory_points=17)
```
